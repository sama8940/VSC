```python
import matplotlib.pyplot as plt

# Data for the bar graph
x = ['A', 'B', 'C', 'D']
y = [10, 5, 12, 8]

# Create the bar graph
plt.bar(x, y)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')

# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# Data for the bar graph
x = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
y = [random.randint(1, 100) for _ in range(len(x))]

# Create the bar graph
plt.bar(x, y)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')

# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'FEH_00200521_240805152851.csv'
df = pd.read_csv(file_path)
df.head()

# Data for the bar graph
x = ["","","","","","",""]
# 
y = []
# 
for index, row in df.iterrows():
    if row.iloc[3] in x:
        y.append(int(row.iloc[5].replace(',','').strip()))


# Create the bar graph
plt.bar(x, y)

# Add labels and title
plt.xlabel('')
plt.ylabel('')
plt.title('')

# Display the graph
plt.show()
```
