```python
import matplotlib.pyplot as plt

# データ
data = [1, 2, 3, 4, 4, 4, 5, 5, 6, 7, 8, 8, 9, 10]

# 階級の指定
bins = [0, 3, 6, 9, 12]

# ヒストグラムの作成
plt.hist(data, bins=bins, edgecolor='black')

# ラベルとタイトルの追加
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Custom Bins')

# ヒストグラムの表示
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'FEH_00200521_240805152851.csv'
df = pd.read_csv(file_path)
df.head()

# 都道府県名のリストを作成
data = []
for index, row in df.iterrows():
    # ○○町だけピックアップする
    if row.iloc[3][-1:] == '町':
        v = row.iloc[5]
        if ( v != '-' ):
            data.append(int(v.replace(',','').strip()))

# Plotting the histogram
plt.hist(data, bins=10, edgecolor='black')

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Displaying the histogram
plt.show()
```
