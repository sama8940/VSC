```python
import requests
# Web APIのエンドポイントURL
url = 'https://api.example.com/data'

# GETリクエストを送信
response = requests.get(url)

# レスポンスのステータスコードを確認
if response.status_code == 200:
    # レスポンスからデータを抽出
    data = response.json()
    print('データの取得に成功しました:', data)
else:
    print('エラー: APIからデータを取得できませんでした')
```

天気予報のデータをフォーマットして表示する。

```python
import requests
import json
# Web APIのエンドポイントURL
url = 'https://www.jma.go.jp/bosai/forecast/data/forecast/130000.json'

# GETリクエストを送信
response = requests.get(url)

# レスポンスのステータスコードを確認
if response.status_code == 200:
    # レスポンスからデータを抽出
    data = response.json()
    formatted_json = json.dumps(data, indent=4)
    print(formatted_json)
else:
    print('エラー: APIからデータを取得できませんでした ')
```

\u9999 のような unicode 表記を通常の漢字表記に直すためには？


```python
unicode_str = "\\u6c17\\u8c61\\u5e81"
# Assuming the Unicode string is stored in the 'unicode_str' variable
normal_str = unicode_str.encode('utf-8').decode('unicode-escape')

print(normal_str)
```

```python
import requests
import json
# Web APIのエンドポイントURL
url = 'https://www.jma.go.jp/bosai/forecast/data/forecast/130000.json'

# GETリクエストを送信
response = requests.get(url)

# レスポンスのステータスコードを確認
if response.status_code == 200:
    # レスポンスからデータを抽出
    data = response.json()
    formatted_json = json.dumps(data, indent=4)
    normal_str = formatted_json.encode('utf-8').decode('unicode-escape')
    print(normal_str)
else:
    print('エラー: APIからデータを取得できませんでした ')
```
